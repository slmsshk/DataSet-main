```python
# Generate the synthetic data without scaling
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=10000, n_features=4, random_state=42)

# Create DataFrames for input and output
input_df = pd.DataFrame(data=X, columns=['Age', 'Weight', 'Expertise', 'Speed'])
output_df = pd.DataFrame(data=y, columns=['Accuracy'])


complete = pd.concat([input_df,output_df],axis=1)

pd.plotting.scatter_matrix(complete)
plt.show()
```

```python
import seaborn as sns

corr = complete.corr()
sns.heatmap(corr,annot=True)
```

```python
X = complete['Age']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x_train,x_test,y_train,y_test =train_test_split(input_df,output_df,test_size=.20,random_state=40)

lr = LinearRegression()
lr.fit(x_train,y_train)

pred = lr.predict(x_test)
print(f'MSE: {int(mean_squared_error(y_test,pred))}')
```
